```python
import pandas as pd

```

```python
# load the dataset
df = pd.read_csv('learningSet.csv', low_memory=False)


# display the head of the dataframe
display(df)
```

Complete the following steps on the CATEGORICAL columns in the dataset:

```python
# create a new dataframe containing only categorical columns from df

categorical = df.select_dtypes(include=['object'])
```

```python
# check for null values in columns
categorical.isnull().sum()
```

```python
# create new empty list
# we will append to this list a set of columns to be dropped later

drop_list = ['OSOURCE', 'ZIP']

# flag columns that have over 85 percent missing values

missing_values = categorical.isnull().sum()
missing_values_percent = (missing_values / len(categorical)) * 100

# Add columns with more than 85% missing values to the drop_list

columns_to_drop = missing_values_percent[missing_values_percent > 85].index.tolist()
drop_list += columns_to_drop

# remove the columns in the drop list from the dataframe
categorical = categorical.drop(columns=drop_list)

display(categorical)
```

```python
# replace missing values with F for gender column

categorical['GENDER'] = categorical['GENDER'].fillna('F')

# reduce the number  of categories in the gender column to M, F and other for all the rest

categorical['GENDER'] = categorical['GENDER'].apply(lambda x: x if x in['M', 'F'] else 'Other' )

display(categorical)
```
