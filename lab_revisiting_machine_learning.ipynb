```python
import pandas as pd

```

```python
# load the dataset
df = pd.read_csv('learningSet.csv', low_memory=False)


# display the head of the dataframe
display(df)
```

```python
# check for null values in columns
df.isnull().sum()
```

```python
# create new empty list
# we will append to this list a set of columns to be dropped later
drop_list = ['OSOURCE', 'ZIP']

# flag columns that have over 85 percent missing values and add them to preceding list
missing_values = df.isnull().sum()
missing_values_percent = (missing_values / len(df)) * 100

# add columns with more than 85 percent missing values to the drop_list
drop_list += missing_values_percent[missing_values_percent > 85].index.tolist()

# remove the columns in the drop list from the dataframe
df = df.drop(columns=drop_list)

# replace missing values with F for gender column
df['GENDER'] = df['GENDER'].fillna('F')

# reduce the number  of categories in the gender column to M, F and other for all the rest

df['GENDER'] = df['GENDER'].apply(lambda x: x if x in['M', 'F'] else 'Other' )

display(df)
```

```python
# just taking a look at the gender column

# display unique values in the GENDER column
print(df['GENDER'].unique())

# display the count of each unique value in the GENDER column
print(df['GENDER'].value_counts())
```
